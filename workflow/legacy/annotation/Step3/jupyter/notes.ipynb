```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 生成模拟数据
np.random.seed(0)  # 为了重现性设置随机种子
x_data = np.random.uniform(0.1, 1, 100)  # 在[0,1]范围内生成100个随机数作为x值
noise = np.random.normal(0.98, 1.02, x_data.size)  # 添加一些噪声
a_true = 0.3  # 假设真实的a值是1.5
# y_data = 2 ** (x_data ** a_true) + noise-1  # 根据公式生成y值

# 定义模型函数
def model_function(x, a):
    return 2 ** ((x ** a+a-1)/a)-1
y_data=model_function(x_data,a_true)
# 使用最小二乘法拟合数据
optimal_parameters, _ = curve_fit(model_function, x_data, y_data, p0=[0.5])

# 提取最佳拟合参数
a_optimal = optimal_parameters[0]

# 使用最佳参数生成模型预测
x_fit = np.linspace(0.1, 1, 100)
y_fit = model_function(x_fit, a_optimal)
rmsd=((y_fit-y_data)**2).mean()
# 绘制数据和拟合曲线
plt.scatter(x_data, y_data, label=f'original data,true a={a_true:.2f}')
plt.plot(x_fit, y_fit, color='red', label=f'fit: a={a_optimal:.2f}, rmsd={rmsd:.2f}')
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100),'--', color='grey', label=f'ref y=x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('fitting $f(x) = 2^{x^a}$-1')
plt.legend()
plt.show()
```
