```python
def fetch_seq(f:str)->str:
    gb:dict=pkl.load(open(f,'rb'))
    seq:str=gb['GBSet']['GBSeq']['GBSeq_sequence']
    return seq
```

```python
import warnings
from Bio import BiopythonWarning
warnings.simplefilter('ignore', BiopythonWarning)
```

```python
from typing import List
trans= lambda x:Seq(x).translate()._data.decode()
def hextranslate(genome:str)->List[str]:
    o=[]
    genome_r=genome[::-1]
    for i in [0,1,2]:
        o.append(trans(genome[i:]))
        o.append(trans(genome_r[i:]))
    return o
```

```python
hextranslate(test_gb['GBSet']['GBSeq']['GBSeq_sequence'])
```

```python
def get_hex(f:str)->str:
    return hextranslate(fetch_seq(f))
```

```python
from glob import glob
from tqdm import tqdm
from pathlib import Path
```

```python

hex_dict={}
with open('get_hex.err','w') as err:
    for i in tqdm(list(Path('genbank_meta/').iterdir())):
        key=i.stem
        try:
            hex_dict[key]=get_hex(i)
        except:
            err.write(key+'\n')

with open('sum_hex.pkl','wb') as f:
    pkl.dump(hex_dict,f)
```
