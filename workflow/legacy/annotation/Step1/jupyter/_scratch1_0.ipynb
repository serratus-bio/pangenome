```python
from Bio import SeqIO
from pathlib import Path
from Bio import Entrez
Entrez.email='zfevan.deng@mail.utoronto.ca'
def fetch_taxonomy(file:Path):
    i=file.stem
    handle=Entrez.efetch(db='nuccore',id=i,rettype="gb", retmode="text")
    s=next(SeqIO.parse(handle,format='genbank'))
    return [i]+s.annotations['taxonomy']
```
